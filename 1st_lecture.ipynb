判別直線をプロットしてくれる関数を作っておく。

```python
def plotline(w,b,plt):
    x1=np.linspace(1,2,2)
    x2=-(b+w[0]*x1)/w[1]
    plt.plot(x1,x2)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
w=np.array([-2,1])
b=2
plotline(w,b,plt)
plt.show()
```

正しく判別しているかどうか計算しよう。
$b+w_0x_0+w_1x_1$の値が正なら$1$、値が負なら$-1$のクラスに分類
これと$y$の積が正なら判別、負なら誤判別

これを計算してくれる予測関数を定義する。

```python
def predict(x,w,b):
    if b+w.dot(x)>0:
        return 1
    else:
        return -1
```

```python
print(predict(x[0],w,b))
print(predict(x[1],w,b))
print(predict(x[2],w,b))
```

完全に分類が終わるまで、自動的に更新し続けるような関数を定義しよう。
パーセプトロン学習器のクラスを作る。

```python
class Perceptron():
    def __init__(self,w,b,n_iter,eta):
        self.w=w
        self.b=b
        self.n_iter=n_iter
        self.eta=eta
    
    def fit(self,X,y):
        for i in range(self.n_iter):
            for j in range(y.shape[0]):
                if self.predict(X[j])==y[j]:
                    pass
                else:
                    self.w,self.b=self.w+self.eta*X[j]*y[j],self.b+self.eta*y[j]
                    
    def predict(self,x):
        if self.b+self.w.dot(x)>0:
            return 1
        else:
            return -1
```

```python
x, y=np.array([[1,1.2],[2,1.8],[1.2,2],[2,1.2],[1.5,1.3]]), np.array([-1,-1,-1,1,1])
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
b=np.random.rand(1)
w=np.random.rand(2)
plotline(w,b,plt)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()

print([predict(x[i],w,b) for i in range(5)])
```

```python
clf=Perceptron(w,b,30,0.1)
clf.fit(x,y)
```

```python
plotline(clf.w,clf.b,plt)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()

print([clf.predict(x[i]) for i in range(5)])
```

課題。
上のクラスを、データのサイズがどのようであっても学習できるように改良しよう。
